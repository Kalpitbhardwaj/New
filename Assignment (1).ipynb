```python
import numpy as np                                      #Importing the Dependencies
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.metrics import accuracy_score
```

# Data Collection and Data Processing

```python

df2=pd.get_dummies(health_data['Fever'])
df3=pd.get_dummies(health_data['Coughing'])
df4=pd.get_dummies(health_data['SoreThroat'])

health_data=pd.concat([df2,health_data],axis=1)
health_data=pd.concat([df3,health_data],axis=1)
health_data=pd.concat([df4,health_data],axis=1)


```

```python
health_data=pd.concat([df,health_data],axis=1)
health_data.drop('Sex',axis=1,inplace=True)
health_data
```

```python

health = health_data.drop(health_data.iloc[:, 15:60], axis = 1)
print(health)
```

```python
X = health.drop(columns = "InitialPCRDiagnosis",axis = 1,inplace=True)
X = health.drop(columns = "CommunityTransmission",axis = 1,inplace=True)
X = health.drop(columns = "Region",axis = 1,inplace=True)
```

```python
health.fillna(0,inplace=True)
```

```python
health.groupby("Diagnosis").mean()
```

# Separating data and labels

```python
X = health.drop(columns = "Diagnosis",axis = 1)      
Y = health["Diagnosis"]
print(X)
print(Y)
```

# Training and Test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.1, stratify = Y, random_state = 1) #Splitting test & train data
```

```python
print(X.shape , X_train.shape , X_test.shape, Y_train.shape)
```

# Model Training --> Logistic Regression

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
X_train_prediction = model.predict(X_train)      #accuracy on training data
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on training data : ', training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)  #accuracy on test data
test_data_accuracy = accuracy_score(X_test_prediction, Y_test) 
```

```python
print('Accuracy on test data : ', test_data_accuracy)
```

# Model Training--> Random Forest Classifier

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train, Y_train)
```

```python
X_test_prediction = model.predict(X_test)                  # accuracy on test data
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on test data : ', test_data_accuracy)
```
